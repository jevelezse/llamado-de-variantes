```python
# Learn about API authentication here: https://plot.ly/python/getting-started
# Find your api_key here: https://plot.ly/settings/api
import plotly.plotly as py
import plotly.graph_objs as go

x = ['Synonymous SNV','Nonsynonymous SNV', 'Stopgain','Stoploss','Unknown']
y = [20046,15237,71,16,427]

data = [
    go.Bar(
        x=x,
        y=y,
        marker=dict(
            color='rgba(0,255,0,0.6)',
            line=dict(
                color='rgba(0,255,0,1.0)',
                width=1.5
            ),
        ),
        opacity=0.6
    )
]
layout = go.Layout(
    annotations=[
        dict(x=xi,y=yi,
             text=str(yi),
             xanchor='center',
             yanchor='bottom',
             showarrow=False,
        ) for xi, yi in zip(x, y)]
) 

fig = go.Figure(data=data, layout=layout)
py.iplot(fig, filename='Variantes_calibradas')
```
